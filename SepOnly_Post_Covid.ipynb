```python
# Import dependencies
import pandas as pd
from pathlib import Path
```

```python
# Store filepath for citibike files in 201509 - 201909  in a variable
file_sep_2021 = Path('Resources/JC-202109-citibike-tripdata.csv')
file_sep_2022 = Path('Resources/JC-202209-citibike-tripdata.csv')
file_sep_2023 = Path('Resources/JC-202309-citibike-tripdata.csv')
```

```python
# Read all September only 2021-2023 data file using pd.read_csv
# and insert one column named yearmonth to depict month and year of the data at column index 1
file_sep_2021_df = pd.read_csv(file_sep_2021)
file_sep_2022_df = pd.read_csv(file_sep_2022)
file_sep_2023_df = pd.read_csv(file_sep_2023)
file_sep_2021_df.insert(1,'yearmonth','Sep2021')
file_sep_2022_df.insert(1,'yearmonth','Sep2022')
file_sep_2023_df.insert(1,'yearmonth','Sep2023')
file_sep_2022_df
```

```python
# Concacendate all the September only data from 2015-2020 along rows, axis=0
sep_post_covid_df = pd.concat([file_sep_2021_df, file_sep_2022_df, file_sep_2023_df], axis=0)
sep_post_covid_df
```

```python
sep_post_covid_df.dtypes
```

```python
# Save output csv file for month of September per Covid citibike trip data
sep_post_covid_df.to_csv('Output/sep_post_covid.csv',index=False, header= True)
```
