```python
# Import dependencies
import pandas as pd
from pathlib import Path
```

```python
# Store filepath for citibike files in 201509 - 201909  in a variable
file_sep_2015 = Path('Resources/JC-201509-citibike-tripdata.csv')
file_sep_2016 = Path('Resources/JC-201609-citibike-tripdata.csv')
file_sep_2017 = Path('Resources/JC-201709-citibike-tripdata.csv')
file_sep_2018 = Path('Resources/JC-201809-citibike-tripdata.csv')
file_sep_2019 = Path('Resources/JC-201909-citibike-tripdata.csv')
file_sep_2020 = Path('Resources/JC-202009-citibike-tripdata.csv')
```

```python
# Read Sep2015 data file using pd.read_csv
file_sep_2015_df = pd.read_csv(file_sep_2015)
file_sep_2016_df = pd.read_csv(file_sep_2016)
file_sep_2017_df = pd.read_csv(file_sep_2017)
file_sep_2018_df = pd.read_csv(file_sep_2018)
file_sep_2019_df = pd.read_csv(file_sep_2019)
file_sep_2020_df = pd.read_csv(file_sep_2020)
# file_sep_2018_df
# file_sep_2018_df.insert(1,'yearmonth','Sep2018')
# file_sep_2018_df
# # file_sep_2018_df.dtypes
```

```python
# Add year and month columns to each citibike trip files in the column 1.
file_sep_2015_df.insert(1,'yearmonth','Sep2015')
file_sep_2016_df.insert(1,'yearmonth','Sep2016')
file_sep_2017_df.insert(1,'yearmonth','Sep2017')
file_sep_2018_df.insert(1,'yearmonth','Sep2018')
file_sep_2019_df.insert(1,'yearmonth','Sep2019')
file_sep_2020_df.insert(1,'yearmonth','Sep2020')
```

```python
# Concacendate all the September only data from 2015-2020 along rows, axis=0
sep_pre_covid_df = pd.concat([file_sep_2015_df, file_sep_2016_df, file_sep_2017_df, file_sep_2018_df, file_sep_2019_df, file_sep_2020_df], axis=0)
sep_pre_covid_df
```

```python
# Save output csv file for month of September per Covid citibike trip data
sep_pre_covid_df.to_csv('Output/sep_pre_covid.csv',index=False, header= True)
```
