```python
# Import dependencies
import pandas as pd
```

```python
# Panda read csv files, month of September only pre Covid citibike ride data
sep_pre_covid_df = pd.read_csv('Output/sep_pre_covid.csv')
pre_covid_df = sep_pre_covid_df.copy()
cleaned_pre_covid_df = pre_covid_df.drop(columns=['tripduration', 'starttime','stoptime','bikeid','usertype','birth year','gender','start station id','end station id'])
cleaned_pre_covid_df
```

```python
# Rename column names on cleaned_pre_covid _df
final_pre_covid_df = cleaned_pre_covid_df.rename(columns={'yearmonth':'month_year','start station name':'start_station_name','start station latitude':'start_station_latitude', 
                                                          'start station longitude':'start_sation_longitude', 'end station name':'end_station_name',
                                                          'end station latitude':'end_station_latitude', 'end station longitude':'end_station_longitude'})
final_pre_covid_df
```

```python
# Panda read csv files, month of September only post Covid citibike ride data
sep_post_covid_df = pd.read_csv('Output/sep_post_covid.csv')
post_covid_df = sep_post_covid_df.copy()
cleaned_post_covid_df = post_covid_df.drop(columns= ['ride_id','rideable_type','started_at','ended_at','start_station_id','end_station_id'])
# cleaned_post_covid_df
renamed_post_covid_df = cleaned_post_covid_df.rename(columns={'yearmonth':'month_year','start_lat':'start_station_latitude',
                                                      'start_lng':'start_sation_longitude','end_lat':'end_station_latitude','end_lng':'end_station_longitude'})
renamed_post_covid_df
```

```python
# Reorder columns on renamed_post_covid_df
final_post_covid_df = renamed_post_covid_df.loc[:,['month_year','start_station_name','start_station_latitude','start_sation_longitude','end_station_name',
                                                   'end_station_latitude','end_station_longitude','member_casual']]
final_post_covid_df
```

```python
# Combine final_pre_covid_df and final_post_covid_df
combined_covid_df = pd.concat([final_pre_covid_df,final_post_covid_df], axis=0)
combined_covid_df
combined_covid_df.value_counts(['start_station_name']).head(10)
```

```python
combined_covid_df.value_counts(['end_station_name']).head(10)
```

```python
# Save output csv file 
combined_covid_df.to_csv('Output/combined_covid.csv',index=False, header= True)
```
